```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://disfold.com/indonesia/companies/?page=1'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find_all('table')
table
```

```python
header = table[0].find_all('th')
header
```

```python
header_titles = [title.text.strip() for title in header]
header_titles
```

```python
import pandas as pd
df = pd.DataFrame(columns=header_titles)
df
```

```python
header_data = table[0].find_all('tr')

for row in header_data[1:]:
    row_data = row.find_all('td')
    row = [td.text.strip() for td in row_data]
    length = len(df)
    df.loc[length] = row

df
```
